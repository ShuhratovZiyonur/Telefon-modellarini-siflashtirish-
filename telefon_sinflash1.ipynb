```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Tasodifiy ma'lumot yaratish (Telefon narxlari va xususiyatlari misol sifatida)
np.random.seed(0)
x = np.random.rand(40, 1) * 10  # Telefon narxlari
y = 2.5 * x ** 3 - 5 * x ** 2 + 3 * x + np.random.randn(40, 1) * 100  # Telefon xususiyatlariga mos ravishda y

# **Bir o'zgaruvchili regressiya**
# Polinom darajasini aniqlash
polynomial_features = PolynomialFeatures(degree=3)
x_poly = polynomial_features.fit_transform(x)

# Model yaratish va o'qitish
model = LinearRegression()
model.fit(x_poly, y)

# Bashorat qilish
y_pred = model.predict(x_poly)

# Natijalarni grafikda chizish
plt.scatter(x, y, color='blue', label='Asl ma\'lumotlar')
plt.plot(x, y_pred, color='red', label='Polinom regressiya')
plt.title('Bir ozgaruvchili Polinom Regressiya')
plt.xlabel('Telefon narxi')
plt.ylabel('Telefon xususiyat(foydali nuqta)')
plt.legend()
plt.grid(True)
plt.show()

# **Ko'p o'zgaruvchili regressiya**
# Qo'shimcha xususiyatlar yaratish
z = np.random.rand(40, 1) * 10  # Telefonning boshqa parametrlari (masalan, kamerasi)
X_multi = np.concatenate((x, z), axis=1)

# Ko'p o'zgaruvchili polinomli regressiya
polynomial_features_multi = PolynomialFeatures(degree=3)
X_poly_multi = polynomial_features_multi.fit_transform(X_multi)

# Model yaratish va o'qitish
model_multi = LinearRegression()
model_multi.fit(X_poly_multi, y)

# Bashorat qilish
y_pred_multi = model_multi.predict(X_poly_multi)

# Mean Squared Error (MSE) baholash
mse = mean_squared_error(y, y_pred_multi)
print("Ko'p o'zgaruvchili Polinom Regressiyaning MSE:", mse)

# Natijalarni tahlil qilish
plt.scatter(range(40), y, color='blue', label='Asl ma\'lumotlar')
plt.plot(range(40), y_pred_multi, color='green', label='Polinom regressiya')
plt.title('Kop ozgaruvchili Polinom Regressiya')
plt.xlabel('Telefon modellarining indeksi')
plt.ylabel('Telefon xususiyatlari')
plt.legend()
plt.grid(True)
plt.show()
```
